```python
#Importing the dependencies
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Data collection and processing
#loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#loading data to Pandas DataFrame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
#print the first five rows of the dataframe
data_frame.head()
```

```python
#adding the target column to the dataframe
data_frame['label'] = breast_cancer_dataset.target
```

```python
#print the first five rows of the dataframe
data_frame.head()
```

```python
#number of rows and columns in dataset
data_frame.shape
```

```python
# getting some information about the data
data_frame.info()
```

```python
#checking for missing values
data_frame.isnull().sum()
```

```python
# finding statistical measures about the data
data_frame.describe()
```

```python
#checking the distribution of target variable(label) 1->B and 0->M
data_frame['label'].value_counts()
```

```python
#grouping the data based on labels
data_frame.groupby('label').mean()
```

Separating features(all the columns except label) and target(label)

```python
#axis=1(when dropping columns) and axis=0(when dropping a row)
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting our data into training and testing data

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
#training the logistic regression model using training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# evaluating the accuracy on training data
X_train_prediction = model.predict(X_train)
#the prediction made by X_train data will be compared with the true values i.e, Y_train data
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on training data = ',training_data_accuracy )
```

```python
# evaluating the accuracy on testing data
X_test_prediction = model.predict(X_test)
#the prediction made by X_test data will be compared with the true values i.e, Y_test data
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on testing data = ',test_data_accuracy )
```
